```python
cols=['MALE SINGER COUNT','FEMALE SINGER COUNT']
df1['TOTAL SINGERS']=df1[cols].sum(axis=1)
df1
```

```python
cols=['MALE SONG COUNT','FEMALE SONG COUNT']
df1['TOTAL SONGS']=df1[cols].sum(axis=1)
df1
```

```python
X=df1.drop('TOTAL SONGS',axis=1)
y=df1['TOTAL SONGS']
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
import category_encoders as ce
```

```python
encoder = ce.OrdinalEncoder(cols=['MOVIE NAME'])
X_train = encoder.fit_transform(X_train)
X_test = encoder.transform(X_test)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor
classifier = RandomForestRegressor(random_state = 42,max_depth=2, n_estimators= 250)
classifier.fit(X_train, Y_train)
```

```python
Y_pred = classifier.predict(X_test)
```

```python
cutoff = 0.7                              
Y_pred_classes = np.zeros_like(Y_pred)    
Y_pred_classes[Y_pred > cutoff] = 1
Y_test_classes = np.zeros_like(Y_pred)
Y_test_classes[Y_test > cutoff] = 1
```

```python
from sklearn.metrics import accuracy_score
(accuracy_score(Y_test_classes,Y_pred_classes))
```
